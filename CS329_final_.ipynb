```python
import pandas as pd

splits = {
    "train": "arabic_dialects_train.csv",
    "validation": "arabic_dialects_dev.csv",
    "test": "arabic_dialects_test.csv"
}

train_df = pd.read_csv("hf://datasets/drelhaj/Arabic-Dialects/" + splits["train"])
val_df   = pd.read_csv("hf://datasets/drelhaj/Arabic-Dialects/" + splits["validation"])
test_df  = pd.read_csv("hf://datasets/drelhaj/Arabic-Dialects/" + splits["test"])

print(train_df.head())
print(train_df.columns)
print(train_df["dialect"].value_counts())
print(train_df.shape, val_df.shape, test_df.shape)
```

```python
print(train_df.head(26))
```

```python
from IPython.display import display
import textwrap

preview = train_df.head(25).copy()

preview["sentence"] = preview["sentence"].apply(
    lambda s: textwrap.fill(str(s), width=55)
)

display(
    preview.style.set_properties(
        subset=["sentence"],
        **{
            "white-space": "pre-wrap",
            "text-align": "right",
            "vertical-align": "top"
        }
    ).set_properties(
        subset=["dialect"],
        **{
            "white-space": "nowrap",
            "vertical-align": "top"
        }
    )
)
```

```python
paste_df = train_df.head(26).copy()
paste_df = paste_df.reset_index().rename(columns={"index": "id"})
paste_df["sentence"] = paste_df["sentence"].astype(str).str.replace(r"\s+", " ", regex=True).str.strip()

print(paste_df.to_csv(sep="\t", index=False))
```
